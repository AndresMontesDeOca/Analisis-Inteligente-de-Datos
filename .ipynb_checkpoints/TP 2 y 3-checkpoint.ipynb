## Trabajo Practico #2 - Transformaciones, Correlacion y PCA

### Ejercicio Numero 2

```python
data = pd.read_excel('Data/chalets.xls', index_col='promotora')

print(data.info())
display(data)
```

#### A - Boxplots

```python
# Se ven valores atipicos en X1, cuando la duracion de la hipoteca es de 37 anos

sns.catplot(kind='box', data=data)
plt.xticks(rotation=45)
plt.show()
```

#### B - Scatterplot

```python
# Todas las variables parecen estar positivamente correlacinadas
# La relacion mas fuerte parece ser X1-X3

sns.pairplot(data=data)
plt.show()
```

#### C - Correlacion

```python
# Vector de medias y matriz de correlacion
# Confirmacion de la alta correlacion entre las variables

print(data.mean(), '\n')
sns.heatmap(data.corr(), vmin=-1, vmax=1, annot=True)
plt.show()
```

#### F.1 - Normalizacion

```python
# Normalizo los datos
from sklearn.preprocessing import StandardScaler
data_std = pd.DataFrame(StandardScaler().fit_transform(data), index=data.index, columns=data.columns)

# Comparo datos normalizados
display(data.describe())
display(data_std.describe())

# La desviacion es de 1.05, toma n-1 como denominador para el std por defecto. Ver ddof
```

#### F.2 - PCA

```python
from sklearn.decomposition import PCA
pca_skl = PCA().fit(data_std)

print('Explicabilidad de cada Componente:', pca_skl.explained_variance_ratio_.round(2))

# Loadings
data_loadings = pd.DataFrame(pca_skl.components_.T, index=data.columns, columns=['PC1', 'PC2', 'PC3'])
display(data_loadings)

# Plot
sns.barplot(data=data_loadings[['PC1']].T)
plt.xticks(rotation=45)
plt.title('PC1')
# plt.show()

# Save PC1 Values
x1 = data_loadings['PC1'].values[0]
x2 = data_loadings['PC1'].values[1]
x3 = data_loadings['PC1'].values[2]
```

#### F.2b - PCA (Libreria Alternativa)

```python
from pca import pca
X = data

pca_alt = pca(n_components=2)
results = pca_alt.fit_transform(X)
print('\n', 'Variance Ratio:', pca_alt.results['variance_ratio'])

# fig, ax = model.biplot(n_feat=4, PC=[0, 1])

pca_alt.plot()
pca_alt.biplot()
```

#### G - Explicabilidad

```python
print('Explicabilidad de cada Componente:', pca_skl.explained_variance_ratio_.round(2))

# Function that calculate the score for each promotora
def calculate_score(promotora):
    return promotora[0]*x1 + promotora[1]*x2 + promotora[2]*x3

# Calculo y muestro el score de cada promotra en PC1, ordenado
promotoras_score = data_std.apply(lambda x: calculate_score(x) , axis=1)
promotoras_score.sort_values(ascending=False)
```

## Test Independencia/Chi2 Pearson

```python
c1 = pd.Series([12, 25], name='No Padece')
c2 = pd.Series([88, 25], name='Padece')
data = pd.DataFrame(index=[c1, c2]).reset_index()
display(data)
print('---------------------------------------------------------------------------------')

Xobs, p_value, _, frecuencias_esperadas = st.chi2_contingency(data)

print('X Observado:', Xobs)
print('P-Value:', p_value)
display(pd.DataFrame(frecuencias_esperadas, columns=data.columns))
```

## Test Homogeneidad

```python
data = sns.load_dataset('titanic')[['class', 'age']].dropna()
display(data.head())

First = data[data['class']=='First']['age']
Second = data[data['class']=='Second']['age']
Third = data[data['class']=='Third']['age']

# Shapiro -> Checkea Normalidad en la Distribucion
print('First:', st.shapiro(First))
print('Second:', st.shapiro(Second))
print('Third:', st.shapiro(Third), '\n')

# Bartletts -> Para Distribuciones Normales
print(st.bartlett(First, Second, Third))

# Levene -> Para Distribuciones No Normales
print(st.levene(First, Second, Third))
```

## Trabajo Practico #3 - Pruebas de Independencia y Homogeneidad


### Ejercicio Numero 2

```python
data_dict = {'F':[14, 159], 'M':[25, 152]}
data_index = ['Twiter', 'No Twiter']
data = pd.DataFrame(data_dict, index=data_index)
display(data)
```

#### A - Plot

```python
fig, ax = plt.subplots()
ax.bar(x=data.index, height=data['F'], label='F')
ax.bar(data.index, data['M'], label='M', bottom=data['F'])

ax.legend()
plt.show()
```

#### B/C - Porcentajes

```python
# Porcentajes
print(data.sum(axis=1)/data.sum().sum(), '\n')
print(data.sum(axis=0)/data.sum().sum())
```

#### D - Frecuencias Esperadas

```python
Xobs, p_value, _, frecuencias_esperadas = st.chi2_contingency(data)

display('Frecuencias Esperadas:', pd.DataFrame(frecuencias_esperadas, columns=data.columns, index=data_index))
display('Frecuencias Observadas:', data)
```

#### E - Diferencia Proporciones

```python
# El P-value obtenido no es tan chico, por ende no puedo rechazar la hipotesis de que las variables son independientes
# Ademas, como vimos anteriormente las proporciones son .49 y .51
print('P-Value:', p_value)
```

### Ejercicio Numero 3

```python
data_dict = {'A':[20, 6, 4, 10], 'B':[18, 22, 6, 19], 'C':[12, 15, 14, 23], 'D':[17, 13, 11, 40], 'E':[67, 56, 35, 92]}
data = pd.DataFrame(data_dict, index=['Norte', 'Sur', 'Este', 'Oeste'])
data
```

#### A/B - Test Homogeneidad

```python
# Son varias muestras para comprar la Homogeneidad de la variable Especialidad

# Test Homogeneidad de Levene
print(st.levene(data.loc['Norte'], data.loc['Sur'], data.loc['Este'], data.loc['Oeste']))

# El P-Value es muy grande, no puedo rechazar H0 que dice que las poblaciones son homogeneas
```
