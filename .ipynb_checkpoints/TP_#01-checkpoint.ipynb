```python
# DataFrames Population

index = pd.Series(['Mariana', 'Maia', 'Sabrina', 'Daniela', 'Alejandra', 'Carla'], name='CandidateName')
features = pd.Series(['Cordialidad', 'Presencia', 'Idioma'])

c1 = pd.Series([80, 80, 90, 80, 70, 90])
c2 = pd.Series([90, 90, 60, 50, 60, 85])
c3 = pd.Series([70, 60, 50, 50, 50, 60])
c4 = pd.Series([60, 65, 70, 70, 55, 80])
c5 = pd.Series([78, 90, 60, 58, 70, 90])
c6 = pd.Series([80, 65, 50, 40, 65, 40])


data_juez1 = pd.concat([index, c1, c2, c3], axis=1).set_index('CandidateName').rename(columns=features)
data_juez2 = pd.concat([index, c4, c5, c6], axis=1).set_index('CandidateName').rename(columns=features)
jueces = [data_juez1, data_juez2]
```

```python
for juez in jueces:
    print(juez.mean(axis=1))
    print('\nTop Performer:', juez.mean(axis=1).idxmax(), '\n')
```

```python
# Concat both DF
data = pd.concat([data_juez1, data_juez2], axis=1)

data.mean(axis=1)
```

```python
# Lets Normalize
from sklearn.preprocessing import StandardScaler, MinMaxScaler


# data_std = pd.DataFrame(MinMaxScaler().fit_transform(data), columns=data.columns, index = data.index)
data_std = pd.DataFrame(StandardScaler().fit_transform(data), columns=data.columns, index = data.index)
display(data_std)

display(data_std.describe().T[['mean', 'std']])

# Poder identificar facilmente lsa debilidades y fortalezas de cada candidata
```

```python
data = pd.read_excel('Data/gorriones.xlsx', index_col='pajaro')

# Trim Column names
data.columns = data.columns.str.replace(' ', '')

print(data.info())
display(data.head())
```

```python
display(data.describe())
```

## Histogramas

```python
# Seteamos series individuales
vivos = data[data.sobrevida==1]
muertos = data[data.sobrevida==-1]

# Preparamos el lienzo
fig = plt.figure(figsize=(10, 12))
for i, col_name in enumerate(data.drop(columns='sobrevida').columns):
    
    # Agregamos cada Axis
    ax = fig.add_subplot(data.shape[1]-1, 1, i+1)
    
    # Ploteamos en cada Axis
    for a in [vivos, muertos]:
        sns.distplot(a[col_name], kde=True, bins=9)
        ax.set_ylabel('')
        ax.set_xlabel('')
        ax.set_title(col_name)
        
fig.tight_layout()  # Improves appearance a bit
plt.show()
```

## PairPlot

```python
sns.pairplot(data, hue='sobrevida')

plt.tight_layout()  # Improves appearance a bit
plt.show()
```

## Correlacion

```python
# Todas las variables muy correlacionadas

sns.heatmap(data.corr(), vmin=-1, vmax=1, cmap='RdYlGn', annot=True)
plt.show()
```
